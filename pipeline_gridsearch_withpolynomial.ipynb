```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                        LinearRegression(**kwargs))
```

## Make some data

```python
import numpy as np
```

```python
def make_data(N,err=1.0,rseed=1):
    # randomly sample the datta
    rng = np.random.RandomState(rseed)
    X = rng.rand(N,1) ** 2
    y = 10 - 1./(X.ravel() + 0.1)
    if err > 0:
        y+=err * rng.randn(N)
        return X , y
```

```python
X,y = make_data(40)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set() # plot formatting
```

```python
X_test = np.linspace(-0.1,1.1,500)[:,None]
```

```python
X_test[0:5]
```

```python
plt.scatter(X.ravel(),y,color='black')
axis=  plt.axis()
```

### fit different polynomials to data and predict

```python
plt.scatter(X.ravel(),y,color='black')
axis=  plt.axis()

for degree in [1,3,5]:
    #print(degree)
    y_test = PolynomialRegression(degree).fit(X,y).predict(X_test)
    plt.plot(X_test.ravel(),y_test,label='degree={0}'.format(degree))
plt.legend(loc='best')
plt.xlim(-0.1,1)
plt.ylim(-0.5,12)
```

### Validation curves

```python
from sklearn.learning_curve import validation_curve
```

```python
degree = np.arange(0,21)
train_score, val_score = validation_curve(PolynomialRegression(),X,y,
                                         'polynomialfeatures__degree',
                                         degree,cv=7)
```

```python
plt.plot(degree, np.median(train_score,1),color='blue',label='training_score')
plt.plot(degree, np.median(val_score,1),color='red',label='validation_score')
plt.legend(loc='best')
plt.ylim(0,1)
plt.xlabel('degree')
plt.ylabel('score')
```

As we should expect: The training score is everwhere higher than the validation score; the training score is monotomically improving with increased model complexity; and the validation score reaches a maximum before dropping off as the model becomes overfit.

From this validation curve we can read off the optimal trade-off between bias and variance; found for a third order polynomial.

The optimal model will generally depend on the size of your training data!!

### Learning curves

```python
from sklearn.learning_curve import learning_curve
```

```python
fig, ax = plt.subplots(1,2,figsize=(16,6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for i, degree in enumerate([2,9]):
    N, train_lc, val_lc = learning_curve(PolynomialRegression(degree),
                                        X,y, cv=7,
                                        train_sizes = np.linspace(0.3,1,25))
    ax[i].plot(N, np.mean(train_lc, 1),color='blue', label = 'training_score')
    ax[i].plot(N, np.mean(val_lc,1), color='red',label='validation_score')
    ax[i].hlines(np.mean([train_lc[-1], val_lc[-1] ]) , N[0],N[-1], color='grey',linestyle='dashed')
    ax[i].set_ylim(0,1)
    ax[i].set_xlabel('training size')
    ax[i].set_ylabel('score')
```

### Grid search

```python
from sklearn.grid_search import GridSearchCV
```

```python
param_grid = {'polynomialfeatures__degree': np.arange(21),
             'linearregression__fit_intercept': [True,False],
             'linearregression__normalize': [True,False]}

grid = GridSearchCV(PolynomialRegression(),param_grid,cv=7)
```

```python
grid.fit(X,y)
```

```python
grid.best_params_
```
